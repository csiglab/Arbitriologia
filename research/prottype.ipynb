```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for the table
data = [
    ["Australia", 26, 65000, 15000, "Large mining exports (iron ore, coal, LNG)"],
    ["Norway", 5.5, 90000, 25000, "Oil and gas exports, sovereign wealth fund supports high GDP"],
    ["Chile", 19, 15000, 2500, "Copper exports dominate"],
    ["Russia", 144, 12000, 2000, "Oil, gas, metals"],
    ["Saudi Arabia", 35, 25000, 20000, "Oil dominant"],
    ["Nigeria", 220, 2500, 200, "Oil exports significant but GDP per capita low"],
    ["Botswana", 2.7, 8000, 3000, "Diamonds significant"],
    ["Venezuela", 28, 7000, 1500, "Oil exports, economic challenges"],
    ["Canada", 39, 55000, 5000, "Oil, minerals, and forest products"]
]

# Column labels
columns = ["Country", "Population (millions)", "GDP per Capita (USD)", 
           "Natural Resource Export Value per Capita (USD)", "Notes"]
```

```python
# Extract GDP and export values
gdp_per_capita = [row[2] for row in data]
export_per_capita = [row[3] for row in data]
countries = [row[0] for row in data]

plt.figure(figsize=(10,7))
plt.scatter(export_per_capita, gdp_per_capita,  color='teal')

# Add labels for each point
for i, country in enumerate(countries):
    plt.text(export_per_capita[i]*1.02, gdp_per_capita[i]*1.02, country, fontsize=9)

plt.xlabel('Natural Resource Export Value per Capita (USD)')
plt.ylabel('GDP per Capita (USD)')
plt.title('Natural Resource Export Value per Capita by Country vs GDP per Capita')
plt.grid(True)
plt.tight_layout()
plt.show()
```
